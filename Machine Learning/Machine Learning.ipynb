# Unsupervised Learning

## The data

scikit-learn allows the creation of sample data sets for different types of ML problems. The following creates a sample data set suited to illustrating k-means clustering.

First, some standard imports and configurations:

```python
import numpy as np
import pandas as pd
import datetime as dt
from pylab import mpl, plt
```

```python
plt.style.use('seaborn')
mpl.rcParams['font.family'] = 'serif'
np.random.seed(1000)
np.set_printoptions(suppress=True, precision=4)
%matplotlib inline
```

Second, the creation of the sample data set.

```python
from sklearn.datasets.samples_generator import make_blobs
```

Creates the sample data set for clustering with 250 samples and 4 centers.

```python
X, y = make_blobs(n_samples=250, centers=4, random_state=500, cluster_std=1.25)
```

```python
plt.figure(figsize=(10, 6))
plt.scatter(X[:, 0], X[:, 1], s=50);
```

# k-means clustering

One of the convenient features of scikit-learn is that it provides a standardized  API to apply different kinds of algorithms. The following code shows the basic steps for k-means clustering that are repeated for other models afterwards:

* Importing the model class
* Instantiating a model object
* Fitting the model object to some data
* Predictingf the outcome given the fitted model for some data

```python
from sklearn.cluster import KMeans   # Imports the model class from scikit-learn.
```

```python
model = KMeans(n_clusters=4, random_state=0)   # Instantiates a model object, given certain parameters; knowledge about the sample data is used to inform the instantiation.
```

```python
model.fit(X)   # Fits the model object to the raw data.
```

```python
y_kmeans = model.predict(X)   # Predicts the cluster (number) given the raw data.
```

```python
y_kmeans[:12]  # Shows some cluster numbers as predicted.
```

```python
plt.figure(figsize=(10,6))
plt.scatter(X[:,0], X[:,1], c=y_kmeans, cmap='coolwarm');
```
