```python
import numpy as np
from pylab import plt, mpl
```

```python
plt.style.use('seaborn')
mpl.rcParams['font.family'] = 'serif'
%matplotlib inline
```

The example function is the following, which is comprised of a trigonometric term and a linear term:

```python
def f(x):
    return np.sin(x) + 0.5 * x
```

The main focus is the approximation of  this function over a given intervalby regression and interpolation techniques. First, a plot of the function to get a better view of what exactly the approximation shall achieve. The interval of interest shall be [-2 pi, 2 pi]. The function create_plot() is a helper function to create the type of plot required.

```python
def create_plot(x, y, styles, labels, axlabels):
    plt.figure(figsize=(10,6))
    for i in range(len(x)):
        plt.plot(x[i], y[i], styles[i], label=labels[i])
        plt.xlabel(axlabels[0])
        plt.ylabel(axlabels[1])
    plt.legend(loc=0)
```

```python
x = np.linspace(-2 * np.pi, 2 * np.pi, 50)
```

```python
create_plot([x], [f(x)], ['b'], ['f(x)'], ['x', 'f(x)'])
```

# Regression

$\hat{Y} = \hat{\beta}_{0} + \sum \limits _{j=1} ^{p} X_{j}\hat{\beta}_{j} $

$\hat{Y} = \hat{\beta}_{0} + \sum \limits _{j=1} ^{p} X_{j}\hat{\beta}_{j} $

```python
res = np.polyfit(x, f(x), deg=1, full=True)         # Linear regression step.
```

```python
res  # Full results: regression parameters, residuals, effective rank
```

```python
ry = np.polyval(res[0], x)            # Evaluation using the regression parameters.
```

```python
create_plot([x, x],[f(x), ry],['b', 'r.'],['f(x)', 'regression'],['x', 'f(x)'],)
```

To account for the sin part of the example function, higher-order monomials are necessary, The next regression attempt takes monomials up to the order of 5 as basis functions. It should not be too surprising that the regression result, as seen now looks much closer to the original function. However, it is still far from being perfect:

```python
reg = np.polyfit(x,f(x), deg=5)
ry = np.polyval(reg, x)
```

```python
create_plot([x, x],[f(x), ry],['b','r.'],['f(x)','regression'],['x','f(x)'])
```

The last attempt takes monomials up to order 7 to approximate the example function.

```python
reg = np.polyfit(x, f(x), 7)
ry = np.polyval(reg, x)
```

```python
np.allclose(f(x), ry)
```

```python
np.mean((f(x) - ry) ** 2)
```

```python
create_plot([x, x], [f(x), ry], ['b', 'r.'], ['f(x)', 'regression'], ['x', 'f(x)'])
```
